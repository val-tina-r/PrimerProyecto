### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Sumamos los precios de todos los productos vendidos en cada una de las tiendas:
VentasTienda1 = sum(tienda['Precio'])
VentasTienda2 = sum(tienda2['Precio'])
VentasTienda3 = sum(tienda3['Precio'])
VentasTienda4 = sum(tienda4['Precio'])

print(f"Las ventas de la tienda 1 son: {VentasTienda1}")
print(f"Las ventas de la tienda 2 son: {VentasTienda2}")
print(f"Las ventas de la tienda 3 son: {VentasTienda3}")
print(f"Las ventas de la tienda 4 son: {VentasTienda4}")
```

```python
#Paleta de colores para las graficas
colores = [plt.colormaps['tab20c'].colors[1], plt.colormaps['tab20c'].colors[5], plt.colormaps['tab20c'].colors[9], plt.colormaps['tab20b'].colors[14]]

#Obtenemos el total de ventas
totalVentas = VentasTienda1 + VentasTienda2 + VentasTienda3 + VentasTienda4

#Hacemos una lista con el equivalente a porcentaje de cada una de las ventas
porcentaje = list(map(lambda x: x*100, [VentasTienda1/totalVentas, VentasTienda2/totalVentas, VentasTienda3/totalVentas, VentasTienda4/totalVentas]))

labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#Gráfico de Pie
plt.pie(porcentaje, labels=labels, colors=colores, autopct='%1.1f%%')
plt.title('Porcentaje de ventas por tienda')
plt.show()
```

# 2. Ventas por categoría

```python
#Obtenemos las ventas totales por categoria para cada tienda agrupando por categoria y sumando el precio de cada producto dentro de la categoria
categoriaTienda1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
categoriaTienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
categoriaTienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
categoriaTienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
```

```python
categoria = categoriaTienda1.index.tolist()
categoria[0] = 'Hogar'
categoria[4] = 'Música'

valores = {'Tienda 1': categoriaTienda1.values.tolist(),
           'Tienda 2': categoriaTienda2.values.tolist(),
           'Tienda 3': categoriaTienda3.values.tolist(),
           'Tienda 4': categoriaTienda4.values.tolist()}

x = np.arange(len(categoria))
width = 0.2
multiplier = 0

fig, ax = plt.subplots(figsize=(10,7),layout='constrained')

for (tienda, valor), color in zip(valores.items(), colores):
    valor = list(map(lambda x: round(x/1e8, 4)*100, valor))
    offset = width * multiplier
    rects = ax.barh(x + offset, valor, height=width, label=tienda, color = color)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_xlabel('Ventas (Millones)')
ax.set_ylabel('Categoría')
ax.set_title('Ventas por categoría')
ax.set_yticks(x + width, categoria)
ax.legend(loc='lower right', ncols=4)

plt.show()
```

### Categoría líder:
- Electrónicos es, con diferencia, la categoría más vendida en todas las tiendas, superando los 400 millones en cada caso.

- La Tienda 4 tiene la menor cifra en esta categoría, mientras que la Tienda 1 lidera con 429,49 millones.

### Categorías con menor rendimiento:

- Libros y Hogar presentan las ventas más bajas en todas las tiendas, rondando entre 8,78 y 15,07 millones.

- Estas categorías aportan poco al total de ingresos y tienen poco margen para aumentar significativamente las ventas sin una estrategia especial.

#### Variaciones entre tiendas:
- En Muebles, la Tienda 3 sobresale con 201,07 millones, frente a los 176,43 millones de la Tienda 2, que tiene el menor valor en esta categoría.

- En Electrodomésticos, la Tienda 1 tiene la cifra más alta (363,69 millones) y la Tienda 4 la más baja (283,26 millones).

- En Música, la Tienda 2 destaca con 104,99 millones, mientras que la Tienda 4 tiene un rendimiento menor (75,1 millones).

### Diversificación de ventas:

- Tiendas como la 1 y la 2 parecen tener un desempeño más equilibrado en varias categorías de alto volumen (Muebles, Electrónicos, Electrodomésticos).

- Tiendas 3 y 4 muestran fortalezas puntuales pero más variación entre categorías.

# 3. Calificación promedio de la tienda


```python
#Obtenemos la calificación promedio de cada tienda haciendo el promedio usando la suma total de las calificaciones sobre el número de calificaciones
calificacionTienda1 = sum(tienda['Calificación'])/len(tienda['Calificación'])
calificacionTienda2 = sum(tienda2['Calificación'])/len(tienda2['Calificación'])
calificacionTienda3 = sum(tienda3['Calificación'])/len(tienda3['Calificación'])
calificacionTienda4 = sum(tienda4['Calificación'])/len(tienda4['Calificación'])

print(f"La calificación promedio de la tienda 1 es: {calificacionTienda1:.2f}")
print(f"La calificación promedio de la tienda 2 es: {calificacionTienda2:.2f}")
print(f"La calificación promedio de la tienda 3 es: {calificacionTienda3:.2f}")
print(f"La calificación promedio de la tienda 4 es: {calificacionTienda4:.2f}")
```

```python
datos = [tienda['Calificación'], tienda2['Calificación'], tienda3['Calificación'], tienda4['Calificación']]

bplot = plt.boxplot(datos, tick_labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], showmeans=True, patch_artist=True,
    medianprops=dict(color='blue', linewidth=2), meanprops=dict(markerfacecolor='yellow', markeredgecolor='black', linewidth=2))

for patch, color in zip(bplot['boxes'], colores):
    patch.set_facecolor(color)

plt.title('Calificación de las tiendas')
plt.ylabel('Calificación')
plt.show()
```

# 6. Análisis del desempeño geográfico

```python
import plotly.express as px

ventas_t1 = tienda.groupby(['lat', 'lon'], as_index=False).agg({'Precio': 'sum'})
ventas_t1['Tienda'] = 'Tienda 1'

ventas_t2 = tienda2.groupby(['lat', 'lon'], as_index=False).agg({'Precio': 'sum'})
ventas_t2['Tienda'] = 'Tienda 2'

ventas_t3 = tienda3.groupby(['lat', 'lon'], as_index=False).agg({'Precio': 'sum'})
ventas_t3['Tienda'] = 'Tienda 3'

ventas_t4 = tienda4.groupby(['lat', 'lon'], as_index=False).agg({'Precio': 'sum'})
ventas_t4['Tienda'] = 'Tienda 4'

ventas_todas = pd.concat([ventas_t1, ventas_t2, ventas_t3, ventas_t4])

fig = px.scatter_geo(
    ventas_todas,
    lat='lat',
    lon='lon',
    size='Precio',
    color='Tienda',
    projection="natural earth",
    scope="south america",
    title='Ventas totales por ubicación - Todas las tiendas',
    size_max=50
)


fig.show()
```

# Informe de análisis para la decisión de venta de tienda - Alura Store

## Introducción
El objetivo de este análisis es ayudar a identificar qué tienda de la cadena Alura Store se debe vender para financiar un nuevo emprendimiento. Para ello, se evaluaron datos de ventas, rendimiento operativo y satisfacción del cliente de las cuatro tiendas, considerando indicadores clave como ingresos totales, categorías de productos más y menos vendidas, calificaciones promedio, productos destacados y coste de envío promedio.

El informe se divide en dos partes:
- Sintesis de los resultados obtenidos del análisis.
- Concluciones y recomendaciones.

## Datos y metodología

El análisis se sustentó en la información proveniente de cada uno de los registros de las tiendas que contenian información sobre:

- Registros históricos de ventas, junto con la clasificación de los productos por categoría.

- Calificación otorgados por los clientes a cada tienda cada vez que se compraba un producto.

- Costos de envío de cada producto.
- Lugar de compra.

Luego, para evaluar el rendimiento de las tiendas, se aplicaron los siguientes pasos:

1. Determinación de ingresos totales para cada tienda y comparación de su peso respecto al total de la cadena.
2. Cálculo de los ingresos por categorías con el fin de identificar cuáles generan mayor aporte a las ventas totales y evaluar su relevancia dentro del portafolio de productos.
3. Cálculo y comparación de las calificaciones promedio para medir la satisfacción del cliente.
4. Identificación de productos con mejor y peor desempeño en cada tienda para evaluar el grado de concentración de las ventas y detectar oportunidades de reemplazo.
5. Análisis del costo promedio de envío como factor de decición.
6. Análisis del desempeño geográfico para poder identificar posibles oportunidades o limitaciones logísticas relacionadas con la ubicación

# Conclusiones y recomendaciones

El análisis integral de ingresos, categorías de productos, satisfacción del cliente, desempeño de ventas por producto, costos de envío y distribución geográfica indica que la Tienda 4 presenta el menor rendimiento general dentro de la cadena, con una participación del 23,6 % en los ingresos totales y sin liderazgo en ninguna categoría clave de productos. Aunque mantiene un nivel de satisfacción del cliente comparable al resto, su portafolio tiene una alta proporción de artículos que también se venden en otras tiendas, lo que reduce el riesgo de pérdida de demanda si la tienda deja de operar. El costo de envío promedio es el más bajo, pero no representa una ventaja competitiva significativa que compense sus menores ingresos.

Se recomienda considerar la venta de la Tienda 4 como opción prioritaria para financiar el nuevo emprendimiento. Esta decisión minimizaría el impacto negativo sobre la cobertura geográfica y la oferta de productos, ya que gran parte de su catálogo más demandado se encuentra disponible en otras tiendas de la cadena. Además, la baja participación en ingresos y la ausencia de categorías dominantes refuerzan su idoneidad como candidata para vender.
